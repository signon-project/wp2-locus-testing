```python
import numpy as np
import pandas as pd
import csv
```

```python
def analyze_csv(file_path):
    df = pd.read_csv(file_path)
    results = {}
    for column in df.columns:
        values = df[column]
        column_min = np.min(values)
        column_max = np.max(values)
        column_avg = np.mean(values)
        results[column] = {
            'min': column_min,
            'max': column_max,
            'avg': column_avg
        }

    return results
```
